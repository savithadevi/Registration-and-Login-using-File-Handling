```python
import hashlib
import re
def signup():
    email = input("Enter email address: ")
    if email_validation(email) == True:
        pwd = input("Enter password: ")
        conf_pwd = input("Confirm password: ")
        if conf_pwd == pwd:
            if password_validation(pwd) == True:
                enc = conf_pwd.encode()
                hash1 = hashlib.md5(enc).hexdigest()
                with open("credentials.txt", "w") as f:
                     f.write(email + "\n") 
                     f.write(hash1)
                f.close()
                print(" \n You have registered successfully! \n ")
            else:
                print(" \n   Password doesn't meet the Conditions \n ")
        else:
            print("Password and Confirm Password doesn't match \n")
    else:
        print(" \n   Enter Valid Email address \n ")


def email_validation(email): # Email Validation Function
    x = re.findall(r"^[a-bA-z]+[0-9a-zA-Z]+\@[a-zA-z]+\.[a-zA-Z]+$",email)
    if x :
        return True
    else:
        return False

def password_validation(password): # Password Validation Function
    special_char = ['$', '@', '#', '%']
    pas = False
    if len(password)>5 and len(password)<16: pas = True
        
    if not any([i.isdigit() for i in password]):
        pas = False
    if not any([i.islower() for i in password]):
        pas = False
    if not any([i.isupper() for i in password]):
        pas = False
    if not any([i.isalpha() for i in password]):
        pas = False
    if not any([i in special_char for i in password]):
        pas = False
        
    return pas


def login():
    email = input("Enter email: ")
    if email_validation(email) == True:
        with open("credentials.txt", "r") as f:
            stored_email, stored_pwd = f.read().split("\n")
        f.close()
        if email == stored_email:    
            pwd = input("Enter password: ")
            auth = pwd.encode()
            auth_hash = hashlib.md5(auth).hexdigest()
            with open("credentials.txt", "r") as f:
                stored_email, stored_pwd = f.read().split("\n")
            f.close()
            if email == stored_email and auth_hash == stored_pwd:
                 print("Logged in Successfully!")
            else:
                 print(" Login failed! \n")
        else:
            print(" \n Email is not Register, Please Signup \n")
    else:
        print(" \n   Enter Valid Email address \n ")


def ForgotPassword():
        email = input("Enter email: ")
        if email_validation(email) == True:
            with open("credentials.txt", "r") as f:
                stored_email, stored_pwd = f.read().split("\n")
            f.close()
            pwd = stored_pwd.encode()
            password = hashlib.md5(pwd).hexdigest()
            passwd=pwd.decode()
            if email == stored_email: 
                print(passwd)
            else:
                print(" \n Email is not Register, Please Signup \n")

        
while 1:
    print("********** Login System **********")
    print("1.Signup")
    print("2.Login")
    print("3.Forgot Password")
    print("4.Exit")
    ch = int(input(" Enter your choice : "))
    if ch == 1:
        signup()
    elif ch == 2:
        login()
    elif ch == 3:
        ForgotPassword()
    elif ch == 4:
        break
    else:
        print(" \n Wrong Choice! \n ")
        
```
